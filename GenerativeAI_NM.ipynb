```python
import pandas as pd

# Read the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/avg-household-size.csv')

# Count of avghouseholdsize
count_avghouseholdsize = data['avghouseholdsize'].value_counts()
print("Count of avghouseholdsize:")
print(count_avghouseholdsize)
```

```python
# Drop rows where avghouseholdsize is less than 2
data_filtered = data[data['avghouseholdsize'] >= 2]
```

```python
# Find sum of duplicated values
sum_duplicated_values = data.duplicated().sum()
print("Sum of duplicated values:")
print(sum_duplicated_values)
```

```python
# Sum of null data
sum_null_data = data.isnull().sum()
print("Sum of null data:")
print(sum_null_data)
```

```python
# Plotting
# Histogram
import matplotlib.pyplot as plt
data['avghouseholdsize'].plot(kind='hist', bins=20)
plt.title('Histogram of Average Household Size')
plt.xlabel('Average Household Size')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart
data['geography'].value_counts().head(10).plot(kind='pie')
plt.title('Top 10 Geographies by Frequency')
plt.ylabel('')
plt.show()
```

```python
# Pie chart comprising only two attributes (avghouseholdsize and countyfips)
data[['avghouseholdsize', 'countyfips']].head(10).plot(kind='pie', subplots=True)
plt.title('avghouseholdsize vs. countyfips')
plt.ylabel('')
plt.show()
```

```python
# Bar graph
data['statefips'].value_counts().plot(kind='bar')
plt.title('Count of State FIPS')
plt.xlabel('State FIPS')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot
data[['avghouseholdsize']].boxplot()
plt.title('Boxplot of Average Household Size')
plt.ylabel('Average Household Size')
plt.show()
```

```python
print("Describe for avghouseholdsize:")
print(data['avghouseholdsize'].describe())
```

```python
# Scatter plot
plt.scatter(data['avghouseholdsize'], data['countyfips'])
plt.title('Scatter plot of Average Household Size vs. County FIPS')
plt.xlabel('Average Household Size')
plt.ylabel('County FIPS')
plt.show()
```

```python
# Boxplot for a particular attribute
data[['avghouseholdsize']].boxplot()
plt.title('Boxplot of Average Household Size')
plt.ylabel('Average Household Size')
plt.show()
```

```python
# Bar graph with groupby
grouped_data = data.groupby('statefips')['avghouseholdsize'].mean()
grouped_data.plot(kind='bar')
plt.title('Average Household Size by State FIPS')
plt.xlabel('State FIPS')
plt.ylabel('Average Household Size')
plt.show()
```
